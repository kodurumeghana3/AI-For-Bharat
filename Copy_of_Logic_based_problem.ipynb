1. Smart Parking Lot Management System


```python
def smart_parking(capacity, logs):
    current_vehicles = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1

        # Track peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_usage)

    if current_vehicles > capacity:
        print("Parking Status: Exceeded Capacity! Alert!")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, vehicle_logs)
```

2. Online Food Delivery Time Estimator


```python
def estimate_delivery_time(distance, traffic, weather):
    # Base time calculation
    base_time = distance * 5   # 5 minutes per km
    total_time = base_time

    # Traffic conditions
    if traffic == "Low":
        total_time += 5
    elif traffic == "Medium":
        total_time += 10
    elif traffic == "High":
        total_time += 20

    # Weather conditions
    if weather == "Clear":
        total_time += 0
    elif weather == "Rainy":
        total_time += 15
    elif weather == "Stormy":
        total_time += 25

    print("Estimated Delivery Time:", total_time, "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

3.Movie Theatre Seat Occupancy Analyzer


```python
def seat_occupancy_analyzer():
    # Input total seats
    total_seats = int(input("Enter Total Seats: "))

    # Input number of booked seats
    booked_count = int(input("Enter Number of Booked Seats: "))

    # Create booked seats list (simulating seat data)
    booked_seats = [1] * booked_count

    # Calculate occupancy percentage
    occupancy = (len(booked_seats) / total_seats) * 100

    print("\n----- Theatre Report -----")
    print("Total Seats:", total_seats)
    print("Booked Seats:", len(booked_seats))
    print("Occupancy:", round(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

    # Suggest additional shows
    if occupancy >= 90:
        print("Suggestion: Consider opening an additional show.")


# Run the function
seat_occupancy_analyzer()
```

4. Cloud Server Load Classification System


```python
def classify_server_load(cpu_readings):
    # Calculate average CPU usage
    total = 0
    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    print("Average CPU Load:", round(average), "%")

    # Classification Rules
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

5. Smart Classroom Resource Usage Monitor


```python
def classroom_resource_monitor(resource_usage):
    overused = []

    # Define maximum allowed usage hours per day
    usage_limit = 8

    # Check each resource
    for resource, hours in resource_usage.items():
        if hours > usage_limit:
            overused.append(resource)

    print("----- Resource Usage Report -----")

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(resource_usage)
```

6.  Online Event Registration Capacity Controller


```python
def event_registration_controller(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0

    # Track registrations using loop
    for i in range(1, total_registrations + 1):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("----- Event Registration Report -----")
    print("Event Capacity:", capacity)
    print("Total Registrations:", total_registrations)
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    # Registration status
    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
capacity = 100
registrations = 105

event_registration_controller(capacity, registrations)
```
